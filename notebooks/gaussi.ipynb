## Voisinages euclidiens

### Définition.

Soient $w \in \mathbb{Z}[i]$ et $z \in \mathbb{C}$. Un **voisinage euclidien** de $z$ par rapport à $(w)$ est l'ensemble 
$B(z,||w||) \cap (w)$
où $||w||$ est la norme géométrique de $w$ et $B(z,||w||)$ le disque ouvert de centre $z$ et de rayon $||w||$.
